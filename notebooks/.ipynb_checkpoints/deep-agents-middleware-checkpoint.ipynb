# Declarative Middleware Composition for LangChain Deep Agents

In LangChain **Deep Agents**, **middleware** is the main structural lever for modular behaviour: you stack independent layers — resilience, observability, governance, context shaping, human gates — and reuse them across products without rewriting the core agent loop. Each layer is a small hook on the model/tool path; combined, they behave like a composable policy pipeline.

That modularity pays off when different teams ship different slices, when you swap implementations per tenant or environment, or when you want the same harness to load different operational profiles. Middleware is where those compositions usually live.

The limitation — and the thread of this notebook — is **how** those compositions are wired today: as one **positional list**. Slot order is part of the contract (outermost versus innermost); the harness does not infer a correct order from the types alone, and nothing checks that separate packages agree when you merge their contributions. So the power of middleware runs into a **sequential** representation that is fragile to reorder, easy to miscompose, and opaque at build time.

What follows: a catalogue of shipped middleware, a short traced run, a clear statement of the ordering problem, then **`MiddlewareStack`** — declarative `slug` / `after` / `before` and `resolve()` — illustrated with five small teaching classes (logging, cache, rate limit, retry, timeout) and optional **LangSmith** screenshots.

## 2. Built-in middleware in `langchain.agents.middleware`

LangChain exposes optional, production-oriented middleware as plain classes (and decorator helpers) you pass in `middleware=[...]`. None of them are required for `create_agent` to run; you add them when you need the capability.

| Class | Role |
| --- | --- |
| `ContextEditingMiddleware` | Prune tool results to manage context size. |
| `FilesystemFileSearchMiddleware` | Glob/Grep search over filesystem files. |
| `HumanInTheLoopMiddleware` | Pause for human approval on selected tool calls. |
| `LLMToolEmulator` | Emulate tools with an LLM instead of executing them. |
| `LLMToolSelectorMiddleware` | Use an LLM to narrow which tools reach the main model. |
| `ModelCallLimitMiddleware` | Cap how many model calls run (per thread / run). |
| `ModelFallbackMiddleware` | Fall back to alternative models on errors. |
| `ModelRetryMiddleware` | Retry failed model calls with backoff. |
| `PIIMiddleware` | Detect / handle PII in conversation traffic. |
| `ShellToolMiddleware` | Register a persistent shell tool (often with `HostExecutionPolicy`, `DockerExecutionPolicy`, or `CodexSandboxExecutionPolicy`). |
| `SummarizationMiddleware` | Summarize history as context limits approach. |
| `TodoListMiddleware` | Todo-list tool and prompts for multi-step work. |
| `ToolCallLimitMiddleware` | Cap tool invocations during a run. |
| `ToolRetryMiddleware` | Retry failed tool calls with backoff. |

Official overview: [Prebuilt middleware](https://docs.langchain.com/oss/python/langchain/middleware/built-in).

The next cell runs a **minimal real agent** with two built-ins (`ModelCallLimitMiddleware`, `TodoListMiddleware`) so you can inspect spans in LangSmith. Enable tracing with `LANGCHAIN_TRACING_V2=true` and your LangSmith key / project env vars as usual.


````python
from langchain.agents import create_agent
from langchain.agents.middleware import ModelCallLimitMiddleware, TodoListMiddleware
from langchain_core.messages import HumanMessage
from langchain_openai import ChatOpenAI

model_name = os.getenv("OPENAI_MODEL", "gpt-4o-mini")
chat = ChatOpenAI(model=model_name, temperature=0, max_tokens=256, api_key=_api_key)

agent_builtin = create_agent(
    model=chat,
    tools=[],
    middleware=[
        ModelCallLimitMiddleware(run_limit=12),
        TodoListMiddleware(),
    ],
)
prompt = (
    "You are planning a tiny change. Use write_todos to record exactly two short tasks, "
    "then reply in one sentence that the plan is recorded."
)
out_b = agent_builtin.invoke({"messages": [HumanMessage(prompt)]})
out_b["messages"][-1].pretty_print()
````

## 3. The ordering problem

`create_agent` only accepts a **positional** `middleware=[...]` list. It does **not** interpret declarative `after` / `before` fields — those are conventions your team encodes in how you **sort** the list. **Index 0 is outermost** (first on the way in, last on the way out); the last index is **innermost**, closest to the model.

The notebook demos below use five **teaching** middleware types (defined later) that mirror common production concerns:

| Layer | What it does |
| --- | --- |
| **Logging** | Records latency, token counts, request IDs |
| **Cache** | Returns a stored response for repeated identical prompts |
| **Rate limit** | Enforces call quotas per time window |
| **Retry** | Re-attempts the call on transient errors |
| **Timeout** | Cancels calls that exceed a wall-clock budget |

Treat that list as a **stack of wrappers**:

```
request ──► [logging] ──► [cache] ──► [rate-limit] ──► [retry] ──► [timeout] ──► model
response ◄─────────────────────────────────────────────────────────────────────────────
```

**Ordering is semantically significant.** A timeout *outside* a retry limits total wall-clock time across all attempts. The same timeout *inside* limits each attempt. Same primitives, different contracts — and the API does not warn you when the order is wrong.

**What the plain list does not give you.** There is no validation that required neighbours exist, that two packages agree on composition, or that implicit constraints hold. Failures tend to show up as subtle runtime behaviour, not at stack construction time.

Illustrative composition with several built-ins (order still matters — not a recommended production ordering):

```python
from langchain.agents import create_agent
from langchain.agents.middleware import (
    ModelCallLimitMiddleware,
    SummarizationMiddleware,
    TodoListMiddleware,
)

agent = create_agent(
    model=chat_model,
    tools=[...],
    middleware=[
        ModelCallLimitMiddleware(run_limit=50),
        TodoListMiddleware(),
        SummarizationMiddleware(chat_model),
    ],
)
```


## 4. Proposal — declarative composition with `MiddlewareStack`

This package adds a **resolve-time** step: each middleware class declares a stable `slug` and optional `after` / `before` constraints. You register instances in any order; `MiddlewareStack.resolve()` performs a topological sort (stable Kahn, tie-break on registration order) and returns the list `create_agent` expects.

### Baseline — positional `middleware`

You choose the **exact list order** passed to `create_agent`. Here: **logging → cache → rate limit → retry → timeout** (outermost → innermost), matching the constraints we will encode declaratively later.

Each class below subclasses LangChain’s **`AgentMiddleware`** (so `create_agent` accepts it) and this package’s **`BaseMiddleware`** (so `MiddlewareStack` can sort it). Hooks use **`wrap_model_call`** — the real interception point for model calls.



````python
import sys

# Demo-only: stdout so Jupyter does not paint INFO as stderr; quiet HTTP client INFO spam.
logging.basicConfig(
    level=logging.INFO,
    format="[mw] %(message)s",
    stream=sys.stdout,
    force=True,
)

from langchain.agents.middleware import AgentMiddleware
from langchain_middleware_stack import BaseMiddleware, MiddlewareStack

mw_log = logging.getLogger("middleware")
mw_log.handlers.clear()
mw_log.addHandler(logging.StreamHandler(sys.stdout))
mw_log.handlers[0].setFormatter(logging.Formatter("[mw] %(message)s"))
mw_log.setLevel(logging.INFO)
mw_log.propagate = False  # keep demo lines on stdout (avoids Jupyter stderr / red)

# AgentMiddleware — LangChain hook surface accepted by create_agent (here: wrap_model_call).
# BaseMiddleware — mixin with defaults for after/before/wires; MiddlewareStack reads slug + edges.


class LoggingMiddleware(AgentMiddleware, BaseMiddleware):
    # Outermost-friendly: observe each model call without changing semantics.
    slug: ClassVar[str] = "logging"
    tools: ClassVar[tuple] = ()  # no extra tools on this class

    def wrap_model_call(self, request, handler):
        # handler(request) runs the next inner layer (eventually the model).
        mw_log.info("[logging] start")
        t0 = time.monotonic()
        try:
            resp = handler(request)
            mw_log.info("[logging] end %.0fms", (time.monotonic() - t0) * 1000)
            return resp
        except Exception as e:
            mw_log.error("[logging] err %s", e)
            raise


class CacheMiddleware(AgentMiddleware, BaseMiddleware):
    # Short-circuit identical ModelRequest message fingerprints (demo-only key).
    slug: ClassVar[str] = "cache"
    after: ClassVar[tuple[str, ...]] = ("logging",)
    tools: ClassVar[tuple] = ()

    def __init__(self):
        self._cache: dict[str, Any] = {}
        self.hits = self.misses = 0

    def wrap_model_call(self, request, handler):
        key = repr([(type(m).__name__, getattr(m, "content", "")) for m in request.messages])
        if key in self._cache:
            self.hits += 1
            mw_log.info("[cache] HIT")
            return self._cache[key]
        self.misses += 1
        mw_log.info("[cache] MISS")
        r = handler(request)
        self._cache[key] = r
        return r


class RateLimitMiddleware(AgentMiddleware, BaseMiddleware):
    # Token-bucket style cap on model calls per sliding time window.
    slug: ClassVar[str] = "rate-limit"
    after: ClassVar[tuple[str, ...]] = ("cache",)
    before: ClassVar[tuple[str, ...]] = ("retry",)
    tools: ClassVar[tuple] = ()

    def __init__(self, max_calls: int = 10, window: float = 1.0):
        self.max_calls, self.window = max_calls, window
        self._calls: list[float] = []

    def wrap_model_call(self, request, handler):
        now = time.monotonic()
        self._calls = [t for t in self._calls if now - t < self.window]
        if len(self._calls) >= self.max_calls:
            raise RuntimeError("rate limit")
        self._calls.append(now)
        mw_log.info("[rl] ok %s/%s", len(self._calls), self.max_calls)
        return handler(request)


class RetryMiddleware(AgentMiddleware, BaseMiddleware):
    # Re-invoke handler on failure; order vs timeout changes total vs per-attempt budgets.
    slug: ClassVar[str] = "retry"
    after: ClassVar[tuple[str, ...]] = ("cache",)
    tools: ClassVar[tuple] = ()

    def __init__(self, max_retries: int = 2, delay: float = 0.0):
        self.max_retries, self.delay = max_retries, delay
        self.total_attempts = 0

    def wrap_model_call(self, request, handler):
        last = None
        for attempt in range(self.max_retries + 1):
            try:
                self.total_attempts += 1
                return handler(request)
            except Exception as e:
                last = e
                mw_log.warning("[retry] %s", e)
                if attempt < self.max_retries:
                    time.sleep(self.delay)
        raise RuntimeError("exhausted") from last


class TimeoutMiddleware(AgentMiddleware, BaseMiddleware):
    # Unix SIGALRM wall-clock bound around the inner handler (demo; not portable).
    slug: ClassVar[str] = "timeout"
    after: ClassVar[tuple[str, ...]] = ("retry",)
    tools: ClassVar[tuple] = ()

    def __init__(self, seconds: int = 60):
        self.seconds = seconds

    def _h(self, s, f):
        raise TimeoutError(self.seconds)

    def wrap_model_call(self, request, handler):
        mw_log.info("[timeout] %ss", self.seconds)
        old = signal.signal(signal.SIGALRM, self._h)
        signal.alarm(self.seconds)
        try:
            return handler(request)
        finally:
            signal.alarm(0)
            signal.signal(signal.SIGALRM, old)


print("AgentMiddleware + BaseMiddleware demo classes ready")
````

````python
from langchain.agents import create_agent
from langchain_core.messages import HumanMessage
from langchain_openai import ChatOpenAI

DEMO_PROMPT = "In one sentence, explain what a middleware pipeline does in a software system."

model_name = os.getenv("OPENAI_MODEL", "gpt-4o-mini")
chat = ChatOpenAI(model=model_name, temperature=0, max_tokens=128, api_key=_api_key)

# Outermost first — this manual ordering is exactly what MiddlewareStack will produce
baseline_layers = [
    LoggingMiddleware(),
    CacheMiddleware(),
    RateLimitMiddleware(30, window=5.0),
    RetryMiddleware(2, delay=0.2),
    TimeoutMiddleware(45),
]
print("Layer order (outermost → innermost):", [m.slug for m in baseline_layers], flush=True)
print(flush=True)

agent = create_agent(model=chat, tools=[], middleware=baseline_layers)
out = agent.invoke({"messages": [HumanMessage(DEMO_PROMPT)]})

final = out["messages"][-1]
content = final.content
text = (
    " ".join(b.get("text", "") for b in content if b.get("type") == "text")
    if isinstance(content, list)
    else content
)
print("\nAgent reply:", flush=True)
print(text.strip() if isinstance(text, str) else text, flush=True)

c = next(m for m in baseline_layers if m.slug == "cache")
r = next(m for m in baseline_layers if m.slug == "retry")
print(
    f"cache → hits={c.hits}  misses={c.misses}  |  retry → attempts={r.total_attempts}",
    flush=True,
)
````

### Why the positional list is still a bottleneck

- **Semantics live in indices** — “retry outside timeout” vs the opposite is a reorder; nothing on the class states the rule.
- **Merging stacks from two packages** — both need a single owner of the final list and out-of-band agreement.
- **`MiddlewareStack`** lets each class declare **`after` / `before`** on **`slug`** values; `resolve()` validates the graph and runs a stable topological sort (Kahn, tie-break on registration order).


### Improved: order from `MiddlewareStack`

The next cell registers the **same five middleware types** as the baseline, but passes them to `stack.add(...)` in a **deliberately wrong order** — `timeout → retry → logging → rate-limit → cache`. That order **breaks** the `after` / `before` rules declared on each class (for example `CacheMiddleware` requires `logging` outside it, and `TimeoutMiddleware` requires `retry` outside it).

**Why that matters:** if you handed that same Python list straight to the agent API, the harness would run middleware **exactly** in that order. Semantics change: logging might not wrap cache hits, timeout and retry nest differently than in the baseline, etc. There is no compile-time error — only wrong runtime behaviour or subtle divergence from the stack you thought you built.

That wrong order simulates real situations: two modules each appending to `middleware`, a refactor that reorders a list, or a merge conflict. **`MiddlewareStack.resolve()`** recomputes the **outermost → innermost** list that satisfies every constraint — here it matches the baseline (`logging → cache → rate-limit → retry → timeout`). The cell prints one compact **order check** block (registered vs target vs after `resolve()`), then runs the agent.


````python
from langchain_core.messages import HumanMessage

model_name = os.getenv("OPENAI_MODEL", "gpt-4o-mini")
chat = ChatOpenAI(model=model_name, temperature=0, max_tokens=128, api_key=_api_key)

log = LoggingMiddleware()
cache = CacheMiddleware()
rlimit = RateLimitMiddleware(30, window=5.0)
retry = RetryMiddleware(2, delay=0.2)
timeout = TimeoutMiddleware(45)

# Same constraints as baseline; 
# this order is WRONG on purpose (does not satisfy after/before).
registration_wrong = [timeout, retry, log, rlimit, cache]
baseline_outer_to_inner = ("logging", "cache", "rate-limit", "retry", "timeout")

stack = MiddlewareStack()
stack.add(registration_wrong)

resolved = stack.resolve()
assert tuple(m.slug for m in resolved) == baseline_outer_to_inner

_reg = [m.slug for m in registration_wrong]
_ok = list(baseline_outer_to_inner)
_fix = [m.slug for m in resolved]
print(
    "Order (outermost → innermost)\n"
    f"  registered (wrong on purpose): {_reg}\n"
    f"  baseline target:               {_ok}\n"
    f"  after resolve():               {_fix}\n",
    flush=True,
)

agent2 = create_agent(model=chat, tools=[], middleware=resolved)
out2 = agent2.invoke({"messages": [HumanMessage(DEMO_PROMPT)]})

final2 = out2["messages"][-1]
content2 = final2.content
text2 = (
    " ".join(b.get("text", "") for b in content2 if b.get("type") == "text")
    if isinstance(content2, list)
    else content2
)
print("\nAgent reply:", flush=True)
print(text2.strip() if isinstance(text2, str) else text2, flush=True)
print(
    f"\ncache → hits={cache.hits}  misses={cache.misses}  |  retry → attempts={retry.total_attempts}\n\n",
    flush=True,
)
````

### Cycle detection

In a declarative stack it is possible to accidentally create circular dependencies: middleware A declares it must come after B, while B declares it must come after A. No valid linear order can satisfy both constraints simultaneously.

`MiddlewareStack` catches this **at build time** — before any agent is created or any request is processed. During the topological sort phase of `resolve()`, if the dependency DAG contains a cycle it raises `MiddlewareCycleError` with the full cycle path. You fix the conflicting constraint immediately, rather than debugging incorrect runtime behaviour that only appears under load.

The example below creates an intentional `ping → pong → ping` cycle:


````python
from langchain_middleware_stack import MiddlewareCycleError


class Ping(BaseMiddleware):
    slug: ClassVar[str] = "ping"
    after: ClassVar[tuple[str, ...]] = ("pong",)


class Pong(BaseMiddleware):
    slug: ClassVar[str] = "pong"
    after: ClassVar[tuple[str, ...]] = ("ping",)


try:
    MiddlewareStack().add(Ping()).add(Pong()).resolve()
except MiddlewareCycleError as e:
    print("cycle:", e)
````

### `wires` — resolve-time attribute injection

As a stack grows, some middleware layers need access to state or callables from other layers. The naïve fix is constructor injection — you pass the dependency when you instantiate the consumer. That works when a single person owns the full stack. It breaks as soon as stacks are assembled across teams or packages: every call site must know the full dependency graph and wire it manually.

`wires` moves the dependency declaration onto the class itself. It maps a local attribute name to `(source_slug, source_attribute)` on another registered instance. After `resolve()` completes the topological sort, it automatically calls `setattr(consumer, local_attr, getattr(provider, source_attr))`:

```python
wires: ClassVar[dict[str, tuple[str, str]]] = {
    "local_attr": ("source_slug", "source_attribute")
}
```

**Wiring is strict** — unlike optional `after` ordering edges, three conditions must hold at `resolve()` time or `MiddlewareWiringError` is raised:
- `source_slug` must be **registered** in the stack
- `source_slug` must appear in this middleware's **`after` tuple** (no hidden DAG edges)
- `source_attribute` must **exist** on the provider instance

Below, `TracingMiddleware` wires its `_ctx` attribute to `AuthMiddleware.trace_context`. No constructor argument needed — the stack injects the reference after resolution:


````python
class AuthMiddleware(BaseMiddleware):
    slug: ClassVar[str] = "auth"

    def __init__(self):
        self.trace_context = {"user": "alice"}

    def wrap(self, handler, *args, **kwargs):
        return handler(*args, **kwargs)


class TracingMiddleware(BaseMiddleware):
    slug: ClassVar[str] = "tracing"
    after: ClassVar[tuple[str, ...]] = ("auth",)
    wires: ClassVar[dict[str, tuple[str, str]]] = {"_ctx": ("auth", "trace_context")}

    def wrap(self, handler, *args, **kwargs):
        mw_log.info("[trace] user=%s", getattr(self, "_ctx", {}).get("user"))
        return handler(*args, **kwargs)


auth_layer, trace_layer = AuthMiddleware(), TracingMiddleware()
MiddlewareStack().add(trace_layer).add(auth_layer).resolve()
print(trace_layer._ctx)
````
