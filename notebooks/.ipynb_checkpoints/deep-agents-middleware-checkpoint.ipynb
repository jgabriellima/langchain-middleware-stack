# Deep Agents with Declarative Middleware
## From Fragile Positional Ordering to Slug-Based DAG Resolution

This notebook demonstrates a real-world problem that emerges when building production-grade deep agent stacks: **middleware ordering is silently fragile**.

We will:
1. Build a mock LLM agent and define four realistic middleware components
2. Show how manual positional ordering fails and produces silent bugs
3. Introduce `langchain-middleware-stack` and show how declarative constraints fix the problem permanently
4. Run an end-to-end agent scenario to confirm correctness

---
### The four middleware components (and why their order matters)

| Middleware | Slug | Constraint | Why |
|---|---|---|---|
| `LoggingMiddleware` | `logging` | *(none)* | Outermost observer — captures everything |
| `CacheMiddleware` | `cache` | `after: ("logging",)` | Must be inside Logging so cache decisions are observed |
| `RetryMiddleware` | `retry` | `after: ("cache",)` | Must be inside Cache — check cache **once**, then retry the model call |
| `TimeoutMiddleware` | `timeout` | `after: ("retry",)` | Must be inside Retry — each retry attempt gets its **own** timeout |

Correct resolved order: **`logging → cache → retry → timeout → model_call`**

## 1. Mock Agent Infrastructure

We build a minimal agent executor so the notebook is self-contained (no real LLM required).  
The `MockLLM` simulates a model that can:
- succeed immediately
- fail a configurable number of times before succeeding (for retry demos)
- simulate latency

The `chain_middleware` helper composes an ordered middleware list into a single callable:  
`chain([A, B, C], model_call)` → `A.wrap(B.wrap(C.wrap(model_call)))`  
so **A is outermost** (runs first / last) and **C is innermost** (closest to the model).

````python
class MockLLM:
    """Simulated language model with configurable failure modes."""

    def __init__(
        self,
        name: str = "gpt-mock",
        fail_times: int = 0,
        latency: float = 0.0,
    ) -> None:
        self.name = name
        self._fail_times = fail_times
        self._call_count = 0
        self._latency = latency

    def call(self, prompt: str) -> str:
        self._call_count += 1
        time.sleep(self._latency)
        if self._call_count <= self._fail_times:
            raise RuntimeError(
                f"[{self.name}] transient error on attempt {self._call_count}"
            )
        return f"[{self.name}] response to: '{prompt}' (attempt #{self._call_count})"

    def reset(self) -> None:
        self._call_count = 0


def chain_middleware(
    ordered: list[Any],
    model_fn: Callable[..., Any],
) -> Callable[..., Any]:
    """Compose middleware list into a single callable.

    ordered[0] is the outermost wrapper (first before-hook, last after-hook).
    model_fn is the innermost call.
    """
    fn = model_fn
    for mw in reversed(ordered):
        inner = fn

        def make_wrapper(middleware: Any, f: Callable[..., Any]) -> Callable[..., Any]:
            def wrapper(*args: Any, **kwargs: Any) -> Any:
                return middleware.wrap(f, *args, **kwargs)
            return wrapper

        fn = make_wrapper(mw, inner)
    return fn


def run_agent(prompt: str, ordered_middleware: list[Any], llm: MockLLM) -> str:
    """Run a single agent turn through the given middleware chain."""
    chain = chain_middleware(ordered_middleware, llm.call)
    return chain(prompt)


print("MockLLM and chain_middleware defined")
````

## 2. Define the Four Middleware Components

Each middleware follows the same interface: a `wrap(handler, *args, **kwargs)` method.  
In this section we write standalone versions — **no ordering constraints declared yet**.  
We will add the declarative constraints in Part 2 when we introduce `MiddlewareStack`.

````python
logger = logging.getLogger("middleware")


class LoggingMiddlewarePlain:
    """Logs entry, exit, and latency of every call."""

    def wrap(self, handler, *args, **kwargs):
        logger.info("[logging] ▶ call start")
        t0 = time.monotonic()
        try:
            result = handler(*args, **kwargs)
            logger.info("[logging] ✔ call end   latency=%.0fms", (time.monotonic() - t0) * 1000)
            return result
        except Exception as exc:
            logger.error("[logging] ✗ call error latency=%.0fms  error=%s", (time.monotonic() - t0) * 1000, exc)
            raise


class CacheMiddlewarePlain:
    """In-memory cache. Returns cached response on hit; calls downstream on miss."""

    def __init__(self) -> None:
        self._cache: dict[str, str] = {}
        self.hits = 0
        self.misses = 0

    def wrap(self, handler, *args, **kwargs):
        key = str(args) + str(kwargs)
        if key in self._cache:
            self.hits += 1
            logger.info("[cache]   ✔ HIT  key=%s", key[:40])
            return self._cache[key]
        self.misses += 1
        logger.info("[cache]   ✗ MISS key=%s", key[:40])
        result = handler(*args, **kwargs)
        self._cache[key] = result
        return result


class RetryMiddlewarePlain:
    """Retries on RuntimeError with fixed delay."""

    def __init__(self, max_retries: int = 2, delay: float = 0.0) -> None:
        self.max_retries = max_retries
        self.delay = delay
        self.total_attempts = 0

    def wrap(self, handler, *args, **kwargs):
        last_exc = None
        for attempt in range(self.max_retries + 1):
            try:
                self.total_attempts += 1
                result = handler(*args, **kwargs)
                if attempt > 0:
                    logger.info("[retry]   ✔ succeeded on attempt %d", attempt + 1)
                return result
            except RuntimeError as exc:
                last_exc = exc
                logger.warning("[retry]   ✗ attempt %d failed: %s", attempt + 1, exc)
                if attempt < self.max_retries:
                    time.sleep(self.delay)
        raise RuntimeError(f"All {self.max_retries + 1} attempts failed") from last_exc


class TimeoutMiddlewarePlain:
    """Per-call timeout (SIGALRM-based, Unix only). Raises TimeoutError if exceeded."""

    def __init__(self, seconds: int = 5) -> None:
        self.seconds = seconds
        self.timeouts_triggered = 0

    def _handler(self, signum, frame):
        self.timeouts_triggered += 1
        raise TimeoutError(f"[timeout] call exceeded {self.seconds}s")

    def wrap(self, handler, *args, **kwargs):
        logger.info("[timeout] ⏱  budget=%ds", self.seconds)
        old = signal.signal(signal.SIGALRM, self._handler)
        signal.alarm(self.seconds)
        try:
            return handler(*args, **kwargs)
        finally:
            signal.alarm(0)
            signal.signal(signal.SIGALRM, old)


print("Four middleware classes defined")
````

---
## Part 1: Without `MiddlewareStack` — The Fragility Problem

When middleware is composed by hand, the developer must know the correct insertion order.  
Here we demonstrate three failure modes:

1. **Silent wrong order** — the stack runs, but semantics are broken
2. **Timeout covers the full retry loop** instead of each individual attempt
3. **Cache is queried on every retry attempt** instead of once before retrying

````python
# ── Correct manual order ─────────────────────────────────────────────────────
# logging → cache → retry → timeout → model
# - Logging wraps everything: sees cache hits, retries, timeouts
# - Cache is checked first: on HIT, neither retry nor timeout run (efficient)
# - Retry wraps timeout: each attempt gets its own per-call timeout

llm_correct = MockLLM("gpt-mock", fail_times=1)  # fails on attempt 1, succeeds on attempt 2

logging_mw = LoggingMiddlewarePlain()
cache_mw   = CacheMiddlewarePlain()
retry_mw   = RetryMiddlewarePlain(max_retries=2)
timeout_mw = TimeoutMiddlewarePlain(seconds=5)

correct_order = [logging_mw, cache_mw, retry_mw, timeout_mw]

print("=" * 60)
print("CORRECT ORDER: logging → cache → retry → timeout")
print("=" * 60)
result = run_agent("What is 2+2?", correct_order, llm_correct)
print(f"\nResult : {result}")
print(f"Cache  : {cache_mw.hits} hits, {cache_mw.misses} misses")
print(f"Retry  : {retry_mw.total_attempts} total handler invocations")
````

````python
# ── Wrong order #1: timeout is outermost ─────────────────────────────────────
# timeout → retry → cache → logging → model
#
# BUG: The timeout budget now covers the ENTIRE retry loop.
# If max_retries=3 and each attempt takes 2s, the 5s timeout fires mid-retry.
# This is not what any developer intends — they want a per-attempt timeout.

llm_wrong1 = MockLLM("gpt-mock", fail_times=1)

logging_mw2 = LoggingMiddlewarePlain()
cache_mw2   = CacheMiddlewarePlain()
retry_mw2   = RetryMiddlewarePlain(max_retries=2)
timeout_mw2 = TimeoutMiddlewarePlain(seconds=5)

wrong_order_1 = [timeout_mw2, retry_mw2, cache_mw2, logging_mw2]  # timeout is FIRST (outermost)

print("=" * 60)
print("WRONG ORDER #1: timeout → retry → cache → logging")
print("BUG: timeout covers the full retry loop, not each attempt")
print("=" * 60)
result = run_agent("What is 2+2?", wrong_order_1, llm_wrong1)
print(f"\nResult : {result}")
print(f"\n⚠ The code RUNS, but the timeout semantics are broken.")
print(f"  Under load (slow model + many retries), the outer timeout")
print(f"  fires before all retry attempts complete — partial failure.")
````

````python
# ── Wrong order #2: cache is innermost ───────────────────────────────────────
# logging → retry → timeout → cache → model
#
# BUG: Cache is checked on EVERY retry attempt.
# On attempt 1: cache MISS → calls model → fails
# On attempt 2: cache MISS again (same key, same empty cache) → calls model again
# The cache never has a chance to short-circuit the retry loop.
#
# This is wasteful (N redundant cache lookups) and semantically wrong:
# the intent is "check cache once, retry the model if missed".

llm_wrong2 = MockLLM("gpt-mock", fail_times=2)

logging_mw3 = LoggingMiddlewarePlain()
cache_mw3   = CacheMiddlewarePlain()
retry_mw3   = RetryMiddlewarePlain(max_retries=3)
timeout_mw3 = TimeoutMiddlewarePlain(seconds=5)

wrong_order_2 = [logging_mw3, retry_mw3, timeout_mw3, cache_mw3]  # cache is LAST (innermost)

print("=" * 60)
print("WRONG ORDER #2: logging → retry → timeout → cache")
print("BUG: cache is inside retry — checked on every attempt")
print("=" * 60)
result = run_agent("What is 2+2?", wrong_order_2, llm_wrong2)
print(f"\nResult : {result}")
print(f"Cache  : {cache_mw3.hits} hits, {cache_mw3.misses} misses")
print(f"Retry  : {retry_mw3.total_attempts} total handler invocations")
print(f"\n⚠ Cache was checked {cache_mw3.misses} times for the same key — redundant.")
print(f"  A cache HIT on a second request also calls the model on each retry.")
````

````python
# ── The fragility test: add a 5th middleware ──────────────────────────────────
# Suppose a new teammate adds RateLimitMiddleware that must run after Cache
# (we only rate-limit actual model calls, not cache hits) and before Retry
# (retry attempts should not re-consume rate-limit quota separately).
#
# Correct position: logging → cache → rate-limit → retry → timeout
#
# In a manual system, the new teammate must:
#   1. Know the current order (undocumented)
#   2. Find the right insertion index
#   3. Pray nobody else changes it

class RateLimitMiddlewarePlain:
    """Limits calls to max_calls_per_window within the time window."""

    def __init__(self, max_calls: int = 10, window: float = 1.0) -> None:
        self.max_calls = max_calls
        self.window = window
        self._calls: list[float] = []
        self.throttled = 0

    def wrap(self, handler, *args, **kwargs):
        now = time.monotonic()
        self._calls = [t for t in self._calls if now - t < self.window]
        if len(self._calls) >= self.max_calls:
            self.throttled += 1
            raise RuntimeError("[rate-limit] quota exceeded")
        self._calls.append(now)
        logger.info("[rate-limit] ✔ allowed (%d/%d in window)", len(self._calls), self.max_calls)
        return handler(*args, **kwargs)


print("Problem: where in this list does RateLimitMiddleware belong?")
print()
print("  [logging, cache, retry, timeout]  ← current")
print()
print("  It must go AFTER cache (only rate-limit real calls, not cache hits)")
print("  and BEFORE retry (each retry attempt should count as one call)")
print()
print("  Correct: [logging, cache, rate-limit, retry, timeout]")
print("  Wrong:   [logging, cache, retry, rate-limit, timeout]  ← rate-limit inside retry")
print("           [logging, rate-limit, cache, retry, timeout]  ← rate-limits cached calls too")
print()
print("In a team of 5+ engineers, this WILL be inserted wrong at some point.")
````

## Part 2: With `MiddlewareStack` — Declarative and Correct

Each middleware now declares its own ordering constraints as class-level attributes.  
`MiddlewareStack` collects them, builds the dependency DAG, and runs a stable topological sort.

**No developer needs to know the correct insertion order.** The constraints are enforced at build time.

We also add `RateLimitMiddleware` and show that it slots in correctly **regardless of when it's added to the stack**.

````python
from langchain_middleware_stack import BaseMiddleware, MiddlewareStack


class LoggingMiddleware(BaseMiddleware):
    """Outermost observer. No ordering constraints — always runs first."""

    slug: ClassVar[str] = "logging"
    # after / before default to () via BaseMiddleware

    def wrap(self, handler, *args, **kwargs):
        logger.info("[logging] ▶ call start")
        t0 = time.monotonic()
        try:
            result = handler(*args, **kwargs)
            logger.info("[logging] ✔ call end   latency=%.0fms", (time.monotonic() - t0) * 1000)
            return result
        except Exception as exc:
            logger.error("[logging] ✗ call error latency=%.0fms  error=%s", (time.monotonic() - t0) * 1000, exc)
            raise


class CacheMiddleware(BaseMiddleware):
    """In-memory cache. Must run inside Logging so cache decisions are observed."""

    slug: ClassVar[str] = "cache"
    after: ClassVar[tuple[str, ...]] = ("logging",)

    def __init__(self) -> None:
        self._cache: dict[str, str] = {}
        self.hits = 0
        self.misses = 0

    def wrap(self, handler, *args, **kwargs):
        key = str(args) + str(kwargs)
        if key in self._cache:
            self.hits += 1
            logger.info("[cache]   ✔ HIT")
            return self._cache[key]
        self.misses += 1
        logger.info("[cache]   ✗ MISS")
        result = handler(*args, **kwargs)
        self._cache[key] = result
        return result


class RateLimitMiddleware(BaseMiddleware):
    """Rate limiter. Must run after Cache (skip cached calls) and before Retry (count each attempt)."""

    slug: ClassVar[str] = "rate-limit"
    after: ClassVar[tuple[str, ...]] = ("cache",)
    before: ClassVar[tuple[str, ...]] = ("retry",)

    def __init__(self, max_calls: int = 10, window: float = 1.0) -> None:
        self.max_calls = max_calls
        self.window = window
        self._calls: list[float] = []
        self.throttled = 0

    def wrap(self, handler, *args, **kwargs):
        now = time.monotonic()
        self._calls = [t for t in self._calls if now - t < self.window]
        if len(self._calls) >= self.max_calls:
            self.throttled += 1
            raise RuntimeError("[rate-limit] quota exceeded")
        self._calls.append(now)
        logger.info("[rate-limit] ✔ allowed (%d/%d)", len(self._calls), self.max_calls)
        return handler(*args, **kwargs)


class RetryMiddleware(BaseMiddleware):
    """Retries on RuntimeError. Must run after Cache so cache is checked before retrying."""

    slug: ClassVar[str] = "retry"
    after: ClassVar[tuple[str, ...]] = ("cache",)

    def __init__(self, max_retries: int = 2, delay: float = 0.0) -> None:
        self.max_retries = max_retries
        self.delay = delay
        self.total_attempts = 0

    def wrap(self, handler, *args, **kwargs):
        last_exc = None
        for attempt in range(self.max_retries + 1):
            try:
                self.total_attempts += 1
                result = handler(*args, **kwargs)
                if attempt > 0:
                    logger.info("[retry]   ✔ succeeded on attempt %d", attempt + 1)
                return result
            except RuntimeError as exc:
                last_exc = exc
                logger.warning("[retry]   ✗ attempt %d failed: %s", attempt + 1, exc)
                if attempt < self.max_retries:
                    time.sleep(self.delay)
        raise RuntimeError(f"All {self.max_retries + 1} attempts failed") from last_exc


class TimeoutMiddleware(BaseMiddleware):
    """Per-attempt timeout. Must run after Retry so each attempt has its own budget."""

    slug: ClassVar[str] = "timeout"
    after: ClassVar[tuple[str, ...]] = ("retry",)

    def __init__(self, seconds: int = 5) -> None:
        self.seconds = seconds
        self.timeouts_triggered = 0

    def _handler(self, signum, frame):
        self.timeouts_triggered += 1
        raise TimeoutError(f"[timeout] call exceeded {self.seconds}s")

    def wrap(self, handler, *args, **kwargs):
        logger.info("[timeout] ⏱  budget=%ds", self.seconds)
        old = signal.signal(signal.SIGALRM, self._handler)
        signal.alarm(self.seconds)
        try:
            return handler(*args, **kwargs)
        finally:
            signal.alarm(0)
            signal.signal(signal.SIGALRM, old)


print("Five declarative middleware classes defined")
````

````python
# ── Resolve in deliberately WRONG insertion order ────────────────────────────
# A developer adds middleware in the wrong order — MiddlewareStack corrects it.

stack = MiddlewareStack()
stack.add(TimeoutMiddleware(seconds=5))   # added first — clearly wrong position
stack.add(RetryMiddleware(max_retries=2))
stack.add(RateLimitMiddleware(max_calls=10))
stack.add(CacheMiddleware())
stack.add(LoggingMiddleware())            # added last — clearly wrong position

ordered = stack.resolve()

print("Insertion order (WRONG):")
print("  ", " → ".join(type(m).__name__ for m in [
    TimeoutMiddleware(), RetryMiddleware(), RateLimitMiddleware(), CacheMiddleware(), LoggingMiddleware()
]))
print()
print("Resolved order (CORRECT):")
print("  ", " → ".join(type(m).__name__ for m in ordered))
print()

# Assert the constraints hold in the resolved order
slugs = [m.slug for m in ordered]
assert slugs.index("cache")      > slugs.index("logging"),    "cache must be after logging"
assert slugs.index("rate-limit") > slugs.index("cache"),      "rate-limit must be after cache"
assert slugs.index("retry")      > slugs.index("rate-limit"), "retry must be after rate-limit"
assert slugs.index("timeout")    > slugs.index("retry"),      "timeout must be after retry"

print("✅ All ordering constraints verified.")
````

````python
# ── MiddlewareStack detects constraint violations at build time ───────────────
from langchain_middleware_stack import MiddlewareCycleError

class BrokenA(BaseMiddleware):
    slug: ClassVar[str] = "broken-a"
    after: ClassVar[tuple[str, ...]] = ("broken-b",)  # A must come after B

class BrokenB(BaseMiddleware):
    slug: ClassVar[str] = "broken-b"
    after: ClassVar[tuple[str, ...]] = ("broken-a",)  # B must come after A  ← CYCLE

try:
    MiddlewareStack().add(BrokenA()).add(BrokenB()).resolve()
    print("ERROR: should have raised")
except MiddlewareCycleError as exc:
    print(f"✅ Cycle detected at resolve() time:")
    print(f"   {exc}")
````

## Cross-Middleware Wiring

`wires` is the fourth primitive. It injects an attribute from one middleware into another **at resolve time**, before any call is made.

Use case: `TracingMiddleware` needs the active trace context from an `AuthMiddleware` to propagate user identity into spans. With `wires`, this injection happens automatically — no manual setter, no global state.

```
consumer.wires = {"_trace_ctx": ("auth", "trace_context")}
```

After `stack.resolve()`, `consumer._trace_ctx` is `auth.trace_context`.

````python
class AuthMiddleware(BaseMiddleware):
    """Injects user identity and exposes a trace context for downstream middleware."""

    slug: ClassVar[str] = "auth"

    def __init__(self) -> None:
        # This attribute will be injected into TracingMiddleware via wires
        self.trace_context = {"user": "alice", "session": "s-001"}

    def wrap(self, handler, *args, **kwargs):
        logger.info("[auth] user=%s", self.trace_context["user"])
        return handler(*args, **kwargs)


class TracingMiddleware(BaseMiddleware):
    """Opens a span for each call. Wires the trace context from AuthMiddleware."""

    slug: ClassVar[str] = "tracing"
    after: ClassVar[tuple[str, ...]] = ("auth",)
    wires: ClassVar[dict[str, tuple[str, str]]] = {
        "_trace_ctx": ("auth", "trace_context"),  # injected at resolve() time
    }

    def wrap(self, handler, *args, **kwargs):
        ctx = getattr(self, "_trace_ctx", {})
        logger.info("[tracing] span.start user=%s session=%s",
                    ctx.get("user", "?"), ctx.get("session", "?"))
        result = handler(*args, **kwargs)
        logger.info("[tracing] span.end")
        return result


auth_mw    = AuthMiddleware()
tracing_mw = TracingMiddleware()

# Before resolve: _trace_ctx does not exist
print(f"Before resolve: hasattr(_trace_ctx) = {hasattr(tracing_mw, '_trace_ctx')}")

wire_stack = MiddlewareStack()
wire_stack.add(tracing_mw)  # added first — wrong order; stack will fix it
wire_stack.add(auth_mw)
wire_stack.resolve()

# After resolve: _trace_ctx is injected
print(f"After resolve:  hasattr(_trace_ctx) = {hasattr(tracing_mw, '_trace_ctx')}")
print(f"Injected value: {tracing_mw._trace_ctx}")
print(f"Same object?    {tracing_mw._trace_ctx is auth_mw.trace_context}")
````

## End-to-End Agent Demo

We now run a realistic multi-turn agent scenario using the full five-middleware stack.

**Scenario:**
- LLM fails on the first attempt (simulates a transient model error)
- Retry recovers on the second attempt  
- A second identical prompt hits the cache — no model call, no retry, no rate-limit consumed

````python
llm = MockLLM("gpt-4o-mock", fail_times=1)  # fails once, then succeeds

log_mw   = LoggingMiddleware()
cache_mw = CacheMiddleware()
rl_mw    = RateLimitMiddleware(max_calls=10)
retry_mw = RetryMiddleware(max_retries=3)
to_mw    = TimeoutMiddleware(seconds=5)

stack = MiddlewareStack()
# Deliberately scrambled insertion order — the stack fixes it
stack.add(to_mw)
stack.add(retry_mw)
stack.add(log_mw)
stack.add(rl_mw)
stack.add(cache_mw)

ordered = stack.resolve()
print("Resolved order:", " → ".join(m.slug for m in ordered))
print()
````

````python
print("=" * 60)
print("TURN 1: fresh prompt — expect cache MISS + 1 retry")
print("=" * 60)
result1 = run_agent("Explain gradient descent", ordered, llm)
print(f"\nResult : {result1}")
print(f"Cache  : {cache_mw.hits} hits, {cache_mw.misses} misses")
print(f"Retry  : {retry_mw.total_attempts} total attempts")
````

````python
print("=" * 60)
print("TURN 2: same prompt — expect cache HIT, zero model calls")
print("=" * 60)
model_calls_before = llm._call_count
result2 = run_agent("Explain gradient descent", ordered, llm)
print(f"\nResult : {result2}")
print(f"Cache  : {cache_mw.hits} hits, {cache_mw.misses} misses")
print(f"Model  : {llm._call_count - model_calls_before} new model call(s) (expected: 0)")

assert result1 == result2, "Cache must return identical result"
assert llm._call_count == model_calls_before, "Cache HIT must not call the model"
print("\n✅ Cache hit verified — model was NOT called on turn 2.")
````

````python
print("=" * 60)
print("TURN 3: new prompt — tests rate limit accounting")
print("=" * 60)
result3 = run_agent("What is backpropagation?", ordered, llm)
print(f"\nResult : {result3}")
print(f"Cache  : {cache_mw.hits} hits, {cache_mw.misses} misses")
print(f"RL     : {len(rl_mw._calls)} call(s) counted in current window")
print()
print("✅ All three turns completed correctly.")
````

## Summary

### What we demonstrated

| Problem | Without `MiddlewareStack` | With `MiddlewareStack` |
|---|---|---|
| Ordering correctness | Developer must know and maintain the right order | Declared once per middleware; enforced at `resolve()` |
| Timeout semantics | Timeout covers whole retry loop if added carelessly | `timeout.after=("retry",)` guarantees per-attempt scope |
| Cache efficiency | Cache queried per retry attempt if inserted wrong | `retry.after=("cache",)` guarantees cache is checked once |
| Adding new middleware | Must find correct insertion index manually | Just `stack.add(new_mw)` — constraints place it automatically |
| Cycles and conflicts | Silent runtime bugs | `MiddlewareCycleError` / `MiddlewareDuplicateSlugError` at build time |
| Cross-middleware state | Global variables or constructor injection | `wires` injects at `resolve()` time, no shared globals |

### The four primitives

```python
class MyMiddleware(BaseMiddleware):
    slug  = "my-middleware"              # stable identity
    after = ("logging",)                 # ordering constraint
    before = ("retry",)                  # ordering constraint
    wires = {"_fn": ("other", "attr")}   # cross-middleware injection
```
